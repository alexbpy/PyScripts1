```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\alexb\spreadex.csv')
```

```python
df
```

```python
# Function to calculate points based on 'POS'
def calculate_points(row):
    if row['POS'] == 'FWD':
        return row['PROJECTED GOALS'] * 3
    elif row['POS'] == 'MID':
        return row['PROJECTED GOALS'] * 4
    else:
        return 0  # Or any default value for other positions

# Apply the function to create the 'POINTS' column
df['POINTS'] = df.apply(calculate_points, axis=1)
```

```python
df
```

```python
df.at[2, 'PROJECTED GOALS'] = 19.25
```

```python
df_mid = df[df['POS']=='MID'].copy()
```

```python
df_att = df[df['POS'] =='FWD'].copy()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pip install adjustText
```

```python
from adjustText import adjust_text 
```

```python
sns.scatterplot(x=df_mid['PRICE'], y=df_mid['POINTS'], color='blue')
sns.regplot(x=df_mid['PRICE'], y=df_mid['POINTS'], scatter_kws={'s': 0}, line_kws={'color': 'red'}, ci=None, label='Trend line')
# Add player names as labels with adjustments to avoid overlap
texts = []
for i, row in df_mid.iterrows():
    if row['POINTS'] >= 22.5:  # Adjust this threshold as needed
        text = plt.text(row['PRICE'], row['POINTS'], row['PLAYER'], ha='left', va='bottom', fontsize=8)
        texts.append(text)

# Adjust the position of text labels to avoid overlap
adjust_text(texts, arrowprops=dict(arrowstyle='-', color='gray'))
plt.xlabel('Price')
plt.ylabel('Proj Points')
plt.title('Scatter Plot with Trend Line')
plt.figure(figsize=(30, 24))

plt.show()
```
