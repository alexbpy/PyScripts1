```python
import matplotlib.pyplot as plt
import mpld3

fig, ax = plt.subplots()

cell_text = [['1', 'Alice', '23'], ['2', 'Bob', '32'], ['3', 'Charlie', '37']]
columns = ('ID', 'Name', 'Age')

table = ax.table(cellText=cell_text, colLabels=columns, loc='center')

table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1, 1.5)

ax.axis('off')

mpld3.show()
```
